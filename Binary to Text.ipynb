```python
a_binary_string = "100011 1110101 1101000 100000 1110011 1101111 100000 1110100 1101000 1100101 100000 1100110 1101111 1110010 1100101 1100011 1100001 1110011 1110100 100000 1101100 1100101 1110110 1100101 1101100 100000 1100001 1101110 1100100 100000 1100010 1101111 1101111 1101011 100000 1101100 1100101 1110110 1100101 1101100 100000 1110011 1101000 1101111 1110101 1101100 1100100 100000 1100001 1101100 1110111 1100001 1111001 1110011 100000 1100010 1100101 100000 1110100 1101000 1100101 100000 1110011 1100001 1101101 1100101 111111"


binary_values = a_binary_string.split()

##Split string on whitespace

ascii_string = ""

for binary_value in binary_values:

    an_integer = int(binary_value, 2)

##Convert to base 2 decimal integer


    ascii_character = chr(an_integer)

##Convert to ASCII character


    ascii_string += ascii_character

##Append character to `ascii_string`


print(ascii_string)
```
