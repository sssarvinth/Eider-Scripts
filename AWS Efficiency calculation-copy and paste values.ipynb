```python
##why hours to ms :-> to narrow down differences between small files to large files(some range from ms to hours, all file can be represented easily as ms

#t-> total audio hours processed by the DA
import datetime
t = '00:00:35.210'
h,m,s = t.split(':' or '.')
sss= str(datetime.timedelta(hours=int(h),minutes=int(m),seconds=float(s)).total_seconds()/60)
print ("total minutes processed:",sss)

##decimal time to seconds
time = 36

hour = int(time)
minutes = (time*60) % 60
seconds = (time*3600) % 60
milliseconds= (time*3600000)%60
total_auto_log_seconds=(3600*hour + 60*minutes+ seconds + 0.001*milliseconds)
total_auto_log_hour= (hour + 0.0166667*minutes+seconds + 2.77778e-7*milliseconds)
#total_auto_log_seconds= (3600*hour + 60*minutes+seconds + milliseconds)
print("total auto log in seconds:",total_auto_log_seconds)

velocity = float(sss)/total_auto_log_hour
print("velocity is the minutes of audio processed per hour:",velocity)
```
