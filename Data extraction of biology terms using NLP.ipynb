```python
import spacy
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
from dframcy import DframCy
from spacy.lang.en.stop_words import STOP_WORDS
nlp = spacy.load('en_core_web_sm')
eider.env.getUploaded('biologydatadump.csv','tmp/biologydatadump.csv')
with open('tmp/biologydatadump.csv',encoding="utf8", errors='ignore') as f:
    doc= nlp(f.read())

##doc_sentences= [sent for sent in doc.sents]
##len(doc_sentences)


dframcy = DframCy(nlp)
spacy_df = dframcy.to_dataframe(doc, ['is_sent_start', 'id', 'text', ]).reset_index()
spacy_df.token_is_sent_start = spacy_df.token_is_sent_start.astype(bool).cumsum() - 1
spacy_df = spacy_df.rename(columns={'token_is_sent_start': 'sentence_id',
                                    'index': 'token_id',
                                    'token_text': 'token_text', })



spacy_df['token_text'].apply(lambda x: [item for item in x if item not in stop])    

with open('tmp/biologydatadump.csv', 'a') as s:
    spacy_df.to_csv(s, header=False)
    
```
