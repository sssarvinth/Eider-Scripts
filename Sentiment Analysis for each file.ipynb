```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
sss= "This is a good file"
negative_sample= "This was the WORST file ever!!!"
positive_sample= "This is the BEST!!! file ever!!!"
print("common example",sid.polarity_scores(sss))
print("negative sample",sid.polarity_scores(negative_sample))
print ("positive sample",sid.polarity_scores(positive_sample))
```
